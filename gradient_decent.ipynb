```
import numpy as np
def gradient_descent(x,y):
  m_curr=1
  c_curr=0
  n=len(x)
  iterations=1000
  learning_rate=0.01
  for i in range(iterations):
    y_p=m_curr*x+c_curr
    cost=sum([val**2 for val in(y-y_p)])
    md=-2*np.sum(x*(y-y_p))
    cd=-2*np.sum(y-y_p)
    m_curr=m_curr-learning_rate*md
    c_curr=c_curr-learning_rate*cd
    print("m={},c={},iteration={},cost={}".format(m_curr,c_curr,i,cost))
x=np.array([0.5,2.3,2.5]) 
y=np.array([1.4,1.9,3.5])  
gradient_descent(x,y) 
```

```
import numpy as np
def gradient_decent1(x,y):
  m_curr=0
  b_curr=0
  n=len(x)
  iterations=200
  learning_rate=0.05
  for i in range(iterations):
    y_p=m_curr*x+b_curr
    cost=(1/n)*sum([val**2 for val in (y-y_p)])
    md=-(2/n)*np.sum(x*(y-y_p))
    bd=-(2/n)*np.sum(y-y_p)
    m_curr=m_curr-learning_rate*md
    b_curr=b_curr-learning_rate*bd
    print("m={},c={},iterations={},cost={}".format(m_curr,b_curr,i,cost))
x=np.array([1,2,3,4,5])
y=np.array([5,7,8,11,15])  
gradient_decent1(x,y)  
```
